# HIERARCHICAL CLUSTERING

IMPORTING LIBRARIES

```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
######
from sklearn import datasets
from scipy.cluster.hierarchy import linkage,dendrogram
```

CREATING BLOBS DATA AND EXTRACTING COLUMNS X1 AND X2

```
x,y=datasets.make_blobs(n_samples=100)
x1=x[:,0]
x2=x[:,1]
```

VISUALIZING DATA

```
plt.scatter(x1,x2)
plt.show()
```

APPLYING NEAREST POINT ALOGORITHM BY USING LINKAGE (google nearest point algorithm)






```
linkage_matrix = linkage(x,"complete")
print(linkage_matrix)
```

CREATING TREE LIKE STRUCTURE (DENDROGRAM)

```
dendrogram = dendrogram(linkage_matrix,truncate_mode=None)
plt.title("Hierarchical Clustering")
plt.show()
```
